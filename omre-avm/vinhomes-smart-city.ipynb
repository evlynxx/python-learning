```python
import pandas as pd
import numpy as np
import openpyxl
```

```python
# lấy tên sheetname (chính là Ngày cập nhật dữ liệu)
path = 'Vinhomes Smart City.xlsx'

wb = openpyxl.load_workbook(path)
sheetnames = list(wb.sheetnames)
updated_date_1 = sheetnames[0]
updated_date_2 = sheetnames[1]
```

### Dữ liệu ngày **18-03-2022**

```python
# Lấy dữ liệu ngày 1
df = pd.read_excel(path, sheet_name=updated_date_1)
df.head()
```

```python
# Insert cột Ngày cập nhật vào df, trong đó chứa tên sheet = ngày đầu
x = list([updated_date_1]*len(df))
df.insert(0, 'Ngày cập nhật', x, True)
df['Ngày cập nhật'] = pd.to_datetime(df['Ngày cập nhật'], format='%d%m%Y')
df
```

```python
col_labels = pd.read_excel(path, sheet_name='Column', header=None) # trong tab column có list các cột cần lấy, biến list các giá trị đó kp header
col_labels = col_labels.loc[0].dropna() # biến list đó thành series -> bỏ giá trị na -> gán lại series vào var
col_labels
```

```python
# chỉ lấy ra các cột trong df thuộc list col_labels
df = df.filter(col_labels)
df
```

```python
# lấy những tên cột mà df chưa có theo list các cột nằm trong col_labels
missing_col = col_labels[~col_labels.isin(df.columns)]
df_null = pd.DataFrame(columns=[missing_col])
df_null
```

```python
# ghép các cột trong df với các cột thiếu ở trên (các cột này sẽ trống)
df_final_1 = pd.concat([df, df_null])
df_final_1 = df_final_1.reindex(col_labels, axis=1) # sắp xếp lại các cột này theo thứ tự các tên cột trong col_labels
df_final_1
```

### Dữ liệu ngày **24-06-2022**

```python
# Lấy dữ liệu ngày 2
df2 = pd.read_excel(path, sheet_name=updated_date_2)
df2.head()
```

```python
# Insert cột ngày vào df, với ngày = tên sheet
y = [updated_date_2]*len(df2)
df2.insert(0, 'Ngày cập nhật', y, True)
df2['Ngày cập nhật'] = pd.to_datetime(df2['Ngày cập nhật'], format='%d%m%Y')
df2
```

Các cột cần lấy của 2 phần dữ liệu là giống nhau, nên chỉ cần 1 bước lấy tên cột ở var col_labels

```python
# lấy những cột cần trong df2
df2 = df2.filter(col_labels)
df2
```

```python
# lấy những cột k có trong df2 nhưng có trong list col_labels
missing_col_2 = col_labels[~col_labels.isin(df2.columns)]
df_null_2 = pd.DataFrame(columns=missing_col_2)
df_null_2
```

```python
# ghép các cột thiếu trong missing_col_2 vào df2
df_final_2 = pd.concat([df2, df_null_2])
df_final_2 = df_final_2.reindex(col_labels, axis=1)
df_final_2
```

```python
# ghép dữ liệu 2 ngày vào với nhau
df_final = pd.concat([df_final_1, df_final_2])
df_final = df_final.reindex(col_labels, axis=1)
df_final
```
