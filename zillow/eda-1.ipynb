```python
import pandas as pd
import numpy as np
import openpyxl
```

```python
sale_price = 'sale-median-sale-price.csv'
list_price = 'sale-median-list-price.csv'
sale_counts = 'sale-counts.csv'
home_value_index = 'zhvi-all-metro.csv'
```

```python
## should write a for loop in here
```

```python
df_sale_price = pd.read_csv(sale_price)
df_sale_price.drop(['SizeRank', 'RegionType'], axis=1, inplace=True)
df_sale_price = df_sale_price.melt(df_sale_price.columns[0:3], var_name='Month', value_name='SalePrice')
df_sale_price
```

```python
df_list_price = pd.read_csv(list_price)
df_list_price.drop(['SizeRank', 'RegionType'], axis=1, inplace=True)
df_list_price = df_list_price.melt(df_list_price.columns[0:3], var_name='Month', value_name='ListPrice')
df_list_price
```

```python
df_sale_counts = pd.read_csv(sale_counts)
df_sale_counts.drop(['SizeRank', 'RegionType'], axis=1, inplace=True)
df_sale_counts = df_sale_counts.melt(df_sale_counts.columns[0:3], var_name='Month', value_name='SaleCounts')
df_sale_counts
```

```python
df_home_value_index = pd.read_csv(home_value_index)
df_home_value_index.drop(['SizeRank', 'RegionType'], axis=1, inplace=True)
df_home_value_index = df_home_value_index.melt(df_home_value_index.columns[0:3], var_name='Month', value_name='SalePrice')
df_home_value_index
```

```python
df_sale_price.to_excel("zillow-data.xlsx", sheet_name='sale-price', encoding='utf-8', index=False)
df_list_price.to_excel("zillow-data.xlsx", sheet_name='list-price', encoding='utf-8', index=False)
df_sale_counts.to_excel("zillow-data.xlsx", sheet_name='sale-counts', encoding='utf-8', index=False)
df_home_value_index.to_excel("zillow-data.xlsx", sheet_name='home_value_index', encoding='utf-8', index=False)
```
